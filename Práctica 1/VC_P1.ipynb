Evento de ratón para mostrar el valor RGB del píxel

```python
# Función que trata eventos de ratón 
def mouse_events(event, x, y, flags, params):
    global px, py, R, G, B
    px = -1
    # Movimiento del puntero     
    if event==cv2.EVENT_MOUSEMOVE:
        #Valores del píxel
        B = frame[y, x, 0]
        G = frame[y, x, 1]
        R = frame[y, x, 2]

        px = x
        py = y
```

```python
vid = cv2.VideoCapture(0)

# Tipografía
font = cv2.FONT_HERSHEY_SIMPLEX
  
lanzado = 0
px = -1
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret: 
        # Activa el manejador en el primer fotograma
        if lanzado == 0:
            # Muestra fotograma
            cv2.imshow('Cam', frame)   
            # Define el nombre del manejador del evento
            cv2.setMouseCallback('Cam', mouse_events)
            lanzado = 1
            
        # Muestra valores RGB intentando centrar en el puntero
        if px > -1:
            cv2.putText(frame, '{}'.format(R), (px-45,py-5), font, 0.5, (0, 0, 255), 1)
            cv2.putText(frame, '     {}'.format(G), (px-54,py-5), font, 0.5, (0, 255, 0), 1)
            cv2.putText(frame, '         {}'.format(B), (px-54,py-5), font, 0.5, (255, 0, 0), 1)
                 
        cv2.imshow('Cam', frame)   
        
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

TAREA:
Pintar círculos en las posiciones del píxel más claro y oscuro de la imagen 

Para encontrar la posición del píxel más claro y oscuro, se puede recorrer la matriz de la imagen en busca de los valores que satisfaga esta condición. El problema es, que si bien funciona, para cada frame relentiza muchísimo la ejecución. Empleando la documentación de OPENCV, se encuentra la función minMaxLoc(), el cual ya nos da las posiciones correspondientes y se pinta en cada frame un circulo en ellos, como una especie de tracking.

Primero, se muestra un ejemplo encontrando la posición del píxel más claro y oscuro de forma manual, recorriendo los píxeles de 5 en 5 para no afectar tanto a la ejecución, entendiendo que no existirán grandes diferencias en cuanto al valor de un píxel y sus 5 píxeles más cercanos en cualquier dirección


```python
vid = cv2.VideoCapture(cv2.CAP_DSHOW)

brightest_pixel = ()
darkest_pixel = ()

while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        h, w, c = frame.shape
    
        max = 0
        min = 255
        for y in range(0, h, 5):
            for x in range(0, w, 5):
                pixel = sum(frame[y, x, 0:3])
                if pixel > max:
                    max = pixel
                    brightest_pixel = (x, y)
                
                elif pixel < min:
                    min = pixel
                    darkest_pixel = (x, y)

        cv2.circle(frame, brightest_pixel, 15, (0,0,0), 2)
        cv2.circle(frame, darkest_pixel, 15, (255,255,255), 2)

        # Muestra fotograma
        cv2.imshow('Tracking pixeles mas claro y oscuro', frame)
        # Detenemos pulsado ESC
        if cv2.waitKey(20) == 27:
            break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

A continuación, se mostrará un ejemplo de la función minMaxLoc() y su uso correspondiente

```python
vid = cv2.VideoCapture(cv2.CAP_DSHOW)
  
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()
  
    if ret:  
        min, max, min_loc, max_loc = cv2.minMaxLoc(cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY))
        cv2.circle(frame, max_loc, 15, (0,0,0), 2)
        cv2.circle(frame, min_loc, 15, (255,255,255), 2)
        
        # Muestra fotograma
        cv2.imshow('Tracking pixeles mas claro y oscuro', frame)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

¿Y si quisieras hacerlo sobre la zona 8x8 más clara/oscura?

```python
vid = cv2.VideoCapture(cv2.CAP_DSHOW)
  
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()
  
    if ret:  

        # Se reescala la imagen dividiendo su ancho y alto entre 8, con la idea de agrupar zonas 8x8 en 1 píxel
        h, w, c = frame.shape
        down_frame = cv2.resize(frame, (int(w/8), int(h/8)), cv2.INTER_NEAREST)

        # Buscamos el píxel más claro y oscuro sobre esta imagen reescalada, que será una aproximación a la zona 8x8 más clara y oscura en
        # la imagen original
        min, max, min_loc, max_loc = cv2.minMaxLoc(cv2.cvtColor(down_frame, cv2.COLOR_RGB2GRAY))

        # Sobre la imagen original, transformamos las coordenadas obtenidas en la imagen reducida para crear círculos sobre
        # los píxeles correspondientes, que en la imagen original será una aproximación a la zona 8x8 más clara y más oscura
        cv2.circle(frame, (max_loc[0]*8+4, max_loc[1]*8+4), 8, (0,0,0), 2)
        cv2.circle(frame, (min_loc[0]*8+4, min_loc[1]*8+4), 8, (255,255,255), 2)
        
        # Muestra fotograma
        cv2.imshow('Tracking zona 8x8 mas clara y oscura', frame)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Pop art
Salida de la webcam inspirada en Andy Warhol

```python
vid = cv2.VideoCapture(0)

#Dimensiones de la cámara
w = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

#Fuerzo a mitad de resolución para ocupar menos pantalla
w=int(w/2)
h=int(h/2)
vid.set(cv2.CAP_PROP_FRAME_WIDTH, w)
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, h)

#Imagen conjunta 2x original
collage = np.zeros((h*2,w*2,3), dtype = np.uint8)
tl = collage[0:h,0:w]
tr = collage[0:h,w:w+w]
bl = collage[h:h+h,0:w]
br = collage[h:h+h,w:w+w]

while True:      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        #Separamos canales
        r = frame[:,:,0]
        g = frame[:,:,1]
        b = frame[:,:,2]

        #Jugamos
        tl[:,:,0] = r
        tl[:,:,1] = g
        tl[:,:,2] = b

        tr[:,:,0] = 255 - g
        tr[:,:,1] = r
        tr[:,:,2] = b
        
        bl[:,:,0] = b
        bl[:,:,1] = 255 - r
        bl[:,:,2] = g

        br[:,:,0] = r
        br[:,:,1] = g
        br[:,:,2] = 255 - b
    
        # Muestra composicion
        cv2.imshow('Cam', collage)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Transformando la imagen capturada en una colección de círculos

```python
vid = cv2.VideoCapture(0)

ncells = 10
off = int(ncells/2)
  
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        #Dimensiones originales
        h, w, c = frame.shape
        #Redimensiono
        down_frame = cv2.resize(frame, (int(w/ncells),int(h/ncells)),cv2.INTER_NEAREST)
        #Dimensiones reducidas
        h2, w2, c2 = down_frame.shape

        #Separamos canales
        r = down_frame[:,:,0]
        g = down_frame[:,:,1]
        b = down_frame[:,:,2]

        # Creamos imagen negra
        gris_up_frame = np.zeros((h2*ncells,w2*ncells,1), dtype = np.uint8)

        for y in range(0,h2):
            for x in range(0,w2):
                # La suma de los valores RGB define el radio del círculo
                rad = int ( (r[y,x] + g[y,x] + b[y,x])/(ncells*3*2))
                cv2.circle(gris_up_frame,(x*ncells+off, y*ncells + off), rad, (255,255,255), -1)

    
        # Muestra fotograma redimensionando a la mitad para que quepa en pantalla
        cv2.imshow('Cam', gris_up_frame)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

TAREA: Haz tu propuesta pop art

En esta propuesta, se realiza un collage 3x2 combinando los ejemplos mostrados y cambiando valores incluso de forma aleatoria. En tres de las seis imágenes se mostrará la imagen transformada a círculos empleando varias combinaciones de colores, y en las otras tres cambiaremos el valor de los planos de forma aleatoria formando así esta propuesta pop art

```python
vid = cv2.VideoCapture(cv2.CAP_DSHOW)

#Dimensiones de la cámara
w = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

#Fuerzo a mitad de resolución para ocupar menos pantalla
w=int(w/2)
h=int(h/2)

vid.set(cv2.CAP_PROP_FRAME_WIDTH, w)
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, h)

#Collage 3x2
collage = np.zeros((h*3,w*2,3), dtype = np.uint8)
tl = collage[0:h,0:w]
tr = collage[0:h,w:w+w]
ml = collage[h:h+h,0:w]
mr = collage[h:h+h,w:w+w]
bl = collage[h+h:h+h+h,0:w]
br = collage[h+h:h+h+h,w:w+w]

while True:      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:

        r = frame[:,:,0]
        g = frame[:,:,1]
        b = frame[:,:,2]

        h2, w2, c2 = frame.shape
        
        #Redimensiono
        down_frame = cv2.resize(frame, (int(w2/10), int(h2/10)), cv2.INTER_NEAREST)

        #Dimensiones reducidas
        h3, w3, c3 = down_frame.shape

        #Separamos canales
        r2 = down_frame[:,:,0]
        g2 = down_frame[:,:,1]
        b2 = down_frame[:,:,2]

        #Jugamos
        tl[:,:,0:3] = 0
        mr[:,:,0:3] = (19, 247, 224)
        bl[:,:,0:3] = 0

        for y in range(0,h3):
            for x in range(0,w3):
                # La suma de los valores RGB define el radio del círculo
                rad = int ( (r2[y,x] + g2[y,x] + b2[y,x])/(10*5))
                cv2.circle(tl,(x*10 + 5, y*10 + 5), rad, (255, 255, 255), -1)
                cv2.circle(mr,(x*10 + 5, y*10 + 5), rad, (247, 19, 217), -1)
                cv2.circle(bl,(x*10 + 5, y*10 + 5), rad, (143, 14, 137), -1)

    
        tr[:,:,0] = int(np.random.randint(100, 255)) - g
        tr[:,:,1] = r
        tr[:,:,2] = b

        ml[:,:,0] = b
        ml[:,:,1] = g
        ml[:,:,2] = int(np.random.randint(100, 255)) - r

        br[:,:,0] = r
        br[:,:,1] = int(np.random.randint(100, 255)) - g
        br[:,:,2] = 255 - b
    
        # Muestra composicion
        cv2.imshow('Pop Art by Raul', collage)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()

```
